```python
# 여러 그래프 나타내기
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2 
y2 = 5*x + 30
y3 = 4*x**2 + 10

plt.plot(x,y1,'bo',label='2*x**2 ') # 특정 형태를 따로 줄 수 있음
plt.plot(x,y2,'r-',label='5*x + 30')
plt.plot(x,y3,'g:',label='4*x**2 + 10')
plt.legend()
plt.show()
```

```python
plt.plot(x,y1,x,y2,x,y3) # 한 줄로 호출 가능하지만, 가독성이 떨어질 수 있다 (위의 방법 권장)
#plt.plot(x,y1,'bo',x,y2,'r-',x,y3,'g:')
plt.show()
```

```python
plt.figure(1) # 인덱스를 지정하면, 지정된 번호의 도면에 그래프가 그려진다 (주의! 넘버링 1부터 시작)
# 설정하지 않으면 디폴트 figure(1)
plt.plot(x,y1)
plt.figure(2) # 새로운 도면에 그려지길 원한다면 figure() 호출
plt.plot(x,y2)
plt.figure(3)
plt.plot(x,y3)
plt.show() # 새로운 도면에 그려진 현재까지의 그래프들을 모두 시각화
```

```python
plt.subplot(2,2,1) # = plt.subplot(221)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.show()
```

```python
x = np.linspace(-4,4,100) # x값 : [-4,4] 범위에서 100개의 값 생성
y1 = x**3
y2 = 10*x**2 - 2

plt.plot(x,y1,x,y2)
plt.show()
```

#산점도(scatter 그래프)
: 두 개의 요소로 이뤄진 데이터 집합의 관계를 시각화 하는 데에 유용하다  
ex. 키와 몸무게, 기온과 아이스크림 판매량의 관계, 공부시간과 시험점수간의 관계  
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter  

```python
height = [165,177,160,180,185,155,172]
weight = [62,67,55,74,90,43,64]

plt.scatter(height,weight) # plot 기존에 쓰던 방법과 동일
plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title("Height & Weight")
plt.grid(True)
plt.show()
```

```python
# 점의 크기 지정 가능

plt.scatter(height,weight,s=500) 
plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title("Height & Weight")
plt.grid(True)
plt.show()
```

```python
# 각각의 사이즈 지정 가능

size = 100*np.arange(1,8)
colors = ['r','g','b','c','m','k','y']
plt.scatter(height,weight,s=size,c=colors)
plt.show()
```

```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 인구 밀도(명/km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

# 위도(latitude)와 경도(longitude)
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]

size = np.array(pop_den) * 0.2  # 마커의 크기 지정 
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 색 지정 

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5) # alpha 투명도 설정
plt.xlabel('경도(lonitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

# 지역명 출력 
for x, y, name in zip(lon, lat, city) : 
  plt.text(x, y, name) 
  #plt.annotate(name,(x,y)) 함수 가능

plt.show()
```

#막대그래프

```python
member_IDs = ['m01','m02','m03','m04']
be_ex = [27,35,40,33]
af_ex = [30,38,42,37]

n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, be_ex, tick_label=member_IDs, width=0.6)
#plt.annotate("o",(-0.3,30))
plt.show()
```

```python
plt.barh(index, be_ex, height=0.6)
plt.show()
```

```python
barWidth = 0.4 # 두 개의 막대가 들어갈 수 있도록 0.4 얇게 조정
plt.bar(index, be_ex, align = 'edge', width = barWidth, label = 'before') # align='edge'로 지정하여 한 쪽으로 치우치도록 설정
plt.bar(index + barWidth, af_ex, align = 'edge', width = barWidth, label = 'after') # bafore막대와 겹치지 않도록 시작

#plt.xticks(index + barWidth, member_IDs) # 두 개의 데이터가 존재할 때에는 tick_label 옵션으로는 변경할 수 없다
#xticks(tick_위치, tick_label) 로 설정
plt.legend()
plt.xlabel('회원ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```

#히스토그램
- 변량(variate) : 자료를 측정하여 나타난 값  
- 계급(class) : 변량을 정해진 간격으로 나눈 구간  
- 도수(frequency) : 나눠진 계급에 속하는 변량의 수  
- 도수 분포표(frequency distribution table) : 계급에 도수를 표시한 표  

> plt.hist(x, [,bins = bins_n 혹은 'auto']) 형태로 작성  
x는 변량데이터, bins는 계습의 개수(default 10), bin='auto' 이면 x에 맞게 자동으로 bins에 값이 입력된다.

```python
scores = [1,1,2,1,2,2,3,2,1,2,3,2,1,2,4,1,2]
plt.hist(scores,bins=4)
plt.grid(True)
plt.show()
```
