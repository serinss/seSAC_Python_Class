#plot 함수
kwargs:가변인자  
plot([x], y, [fmt], *, data=None, **kwargs)  
각각 x축과 y축의 좌표값을 의미  
plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

```python
# 선형 함수에 대한 정의와 그래프 시각화
import numpy as np
data = np.arange(10) # y축 데이터
data
```

```python
plt.plot(data) # 도면에 추가, x축 데이터는 0부터 순차적으로 자동 부여
plt.show() # 시각화
# x축 눈금값 = xticks -> [] 범위값 직접 설정해줄 수 있음
```

```python
plt.plot([1,4,7,9,12,15]) # y값만 인자로 전달, x축 데이터는 0부터 순차적으로 자동 부여
plt.show()
```

```python
#제목 부여하기
plt.plot([1,4,7,9,12,15])
plt.title("plt.plot([1,4,7,9,12,15])")
plt.show()
```

```python
#x,y 전달
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]) #[x값],[y값] -> (1,1),(2,2),(3,9),(4,7),(5,12),(7,15) 를 나타내었다
plt.title("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15])")
plt.show()
```

```python
# 데이터를 두 개 이상 전달하기
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo',[1,2,3,4,5,7],[8,9,10,15,10,11],'r^')
plt.title("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo',[1,2,3,4,5,7],[8,9,10,15,10,11],'r^')")
plt.show()
```

```python
# 데이터를 두 개 이상 전달하기
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo')
plt.plot([1,2,3,4,5,7],[8,9,10,15,10,11],'r^') # 가독성을 위해 plt 함수 두 번 호출 가능
plt.title("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo'),plt.plot([1,2,3,4,5,7],[8,9,10,15,10,11],'r^')")
plt.show()
```

```python
# 축 범위 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,81,87,78,94],'r^')
plt.axis([0,6,50,100]) #[x축 범위,y축 범위]
plt.title("scores")
plt.show()
```

```python
# 축 이름 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo',label='class 1')
plt.plot([1,2,3,4,5],[95,81,87,78,94],'r^',label='class 2')
plt.axis([0,6,50,100]) #[x축 범위,y축 범위]
plt.title("scores") # 그래프 이름 설정
plt.xlabel('number') # 축 이름 설정
plt.ylabel('score')
plt.legend() # 범례(label) 표시하기 / loc 로 위치 지정 가능
plt.show()
```

```python
# sub 도표 사용하기
plt.subplot(221) # 2x2 도표 중에 1번째
plt.title("221")
plt.plot([1,2,3,4],'ro-')

plt.subplot(222) # 2x2 도표 중에 2번째
plt.title("222")
plt.plot([4,3,2,1],'b.:')

plt.subplot(223) # 2x2 도표 중에 3번째
plt.title("223")
plt.plot([1,3,2,4],'c^-.')

plt.subplot(224) # 2x2 도표 중에 4번째
plt.title("224")
plt.plot([2,4,1,3],'k.--')

plt.show()
```

```python
# y축에 표현할 값을 지정 + 디폴트 색상 확인
ys_str=[]
for i in range(1,31):
  plt.plot([i,i]) #수평선
  
  if i%3 == 0:
    ys_str.append(str(i)) # 3의 배수인 것은 y축에 숫자 표현
  else:
    ys_str.append(" ") # 아닌 것들은 공백
plt.title("default color, set yticks")
plt.yticks(range(1,31),ys_str)
plt.show()
```

```python
#배경색
plt.plot([1,1],'ro-')
ax = plt.gca()
ax.set_facecolor('violet')
plt.title("set background color")
plt.show()
```

```python
# data 사용하기
mdic = dict()
mdic['hour'] = [0,1,2,3,4,5]
mdic['distance'] = [0,5,20,45,70,95]

plt.plot('hour','distance',data=mdic)
plt.title("using mdic data")
plt.savefig('demo.png',dpi=400,bbox_inches='tight') #해상도 설정하여 해당 이미지 저장하기
plt.show()
```

#막대그래프
1. 수직 막대그래프 : matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)   
2. 수평 막대그래프 : matplotlib.pyplot.barh(y, width, height=0.8, left=None, *, align='center', **kwargs) -> horizental

```python
names = ['a','b','c']
scores = [80,90,85]

plt.bar(names,scores) # 막대 그래프
plt.xlabel('name')
plt.ylabel('score')
plt.title('name-score bar')
plt.show()

plt.plot(names,scores) # 꺽쇠 그래프
plt.xlabel('name')
plt.ylabel('score')
plt.title('name-score bar')
plt.show()

# 어떠한 그래프가 적절한지는 판단해아함
```

#파이그래프
matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, data=None)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html?highlight=pie#matplotlib.pyplot.pie

```python
locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
plt.pie(incomes,labels=locals)
plt.title("지역별 수입")
plt.show()
```

```python
# 파이그래프 - 강조하고 싶은 값 돌출시키기

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0] # 기본 원의 반지름 = 1 , 0~1 사이의 값으로 어느정도 돌출시킬 것인지 설정
plt.pie(incomes,labels=locals,explode=exs)
plt.title("지역별 수입 - 돌출시키기")
plt.show()
```

```python
# 명확히 퍼센트 표시하기

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels=locals,explode=exs,autopct='%.2f%%')
plt.title("지역별 수입 - 퍼센트 표시하기")
plt.show()
```

```python
# 퍼센트 표시할 위치 지정하기

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels=locals,explode=exs,autopct='%.2f%%',pctdistance=0.7)
plt.title("지역별 수입 - 퍼센트 위지 지정하기")
plt.show()
```

```python
# 그림자 효과

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels=locals,explode=exs,autopct='%.2f%%',pctdistance=0.7,shadow=True)
plt.title("지역별 수입 - 그림자 효과")
plt.show()
```

```python
#label 위치 지정

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels=locals,explode=exs,autopct='%.2f%%',pctdistance=0.7,shadow=True,labeldistance=1.2) #default = 1.1
plt.title("지역별 수입 - 라벨 표시 위지 지정하기")
plt.show()
```

```python
# 데이터값의 시작은 디폴트 3시 설정되어 있음 / 방향은 역시계 방향
# 시계 방향으로 각도 계산하기

locals = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels=locals,explode=exs,autopct='%.2f%%',startangle=90,counterclock=False) # 12시부터 시작하도록 설정 / 시계 방향으로 설정
plt.title("지역별 수입 - 시계 방향으로 각도 계산하기")
plt.show()
```

#주석 달기
matplotlib.pyplot.annotate(text, xy, *args, **kwargs)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html?highlight=annotate#matplotlib.pyplot.annotate

```python
#주석 달기
#annotation(lineType)은 점 (0.3,0.3)과 (0.4,0.4) 사이의 디폴트 위치에 주석을 만듭니다.

colors = ['red','green','blue']
names = ['a','b','c']
income = [200,230,300]
plt.bar(names,income,color=colors)
plt.annotate('a is eh',(-0.3,220)) # a 가 있는 위치 = 0 보다 왼쪽이므로 -0.3 으로 지정
plt.title("색상 지정")
plt.show()
```
